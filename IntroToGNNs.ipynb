## Multiple Edge Attribiutes

Create adjacency matrices with multiple edge attributes witg structured dtypes

```python
import numpy as np

G = nx.Graph()
G.add_edge(0, 1, weight = 10)
G.add_edge(1, 2, cost = 5)
G.add_edge(2, 3, weight = 3, cost = -4.0)
dtype = np.dtype([('weight', int), ('cost', float)])
A = nx.to_numpy_array(G, dtype = dtype, weight = None)
print('weight ----------')
print(A['weight'])
print('cost ---------')
print(A['cost'])
nx.draw(G)
print(G.edges)
print(G.nodes)
```

Returning adjacency matrix of a dictionary of dictionaries

```python
G = nx.Graph()
G.add_edges_from(
    [
        (0, 1, {'weight': 1.0}),
        (1, 2, {'weight': 2.0}),
        (2, 0, {'weight': 1.0})
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print('-------------------')
d = nx.to_dict_of_dicts(G)
print(d)
print(d[1][2]['weight'])
```

```python
G = nx.MultiGraph()
G.add_edge(0, 1, key = 'a', weight = 1.0)
G.add_edge(0, 1, key = 'b', weight = 5.0)
nx.draw(G)
print(G.edges)
print(G.nodes)
print('-------------------')
d = nx.to_dict_of_dicts(G)
print(d)
print(d[0][1]['b']['weight'])
```

```python
G = nx.Graph(
    [
        ('A', 'B', {'cost': 1, 'weight': 7}),
        ('C', 'E', {'cost': 9, 'weight': 10})
    ]
)
nx.draw(G)
print(G.nodes)
print(G.edges)
print('---------------')
df = nx.to_pandas_edgelist(G)
print(df)
print('---------------')
df = nx.to_pandas_edgelist(G, nodelist=['B', 'E'])
print(df)
print('--------------------')
df = nx.to_pandas_edgelist(G, nodelist= ['A','C'])
print(df)
print('-------------------')
df[['source', 'target', 'cost', 'weight']]
print(nx.adjacency_matrix(G).todense())
```

```python
G = nx.MultiGraph([('A', 'B', {'cost': 7}), ('A', 'B', {'cost': 9})])
nx.draw(G)

print(G.edges)
print(G.nodes)
print('-----------------')
df = nx.to_pandas_edgelist(G, nodelist=['A', 'C'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
print('-----------------')
df = nx.to_pandas_edgelist(G, nodelist=['B'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
A = nx.adjacency_matrix(G)
print(A.todense())
```
